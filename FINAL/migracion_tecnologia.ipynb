```python
def plot_pie(valores, etiquetas):
    fig1, ax1 = plt.subplots()
    ax1.pie(valores, labels=etiquetas, autopct='%1.1f%%',
            shadow=True, startangle=90)
    ax1.axis('equal')
    plt.show()
```

```python
def plot_bar(items_x, valores, etiquetas, limites, rotation="vertical"):
    fig, ax = plt.subplots()
    plt.bar(items_x, valores)
    plt.xticks(items_x, etiquetas)
    plt.ylim(*limites)
    plt.xticks(rotation=rotation)
    plt.show()
```

    * Respuesta a: ¿Como se ve la migración en Ecuador?

```python
entrada_migration_2017 = migration_2017[migration_2017["tip_movi"] == "Entrada"]
serie_migration_2017 = entrada_migration_2017['edad'].groupby(entrada_migration_2017['sex_migr']).agg('count')

df_agrupado_edad_genero = pd.DataFrame({'genero': serie_migration_2017.index, 'edad': serie_migration_2017.data})

print(df_agrupado_edad_genero)

fig1, ax1 = plt.subplots()
ax1.pie(df_agrupado_edad_genero['edad'], labels=df_agrupado_edad_genero['genero'], autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
entrada_migration_2017 = migration_2017[(migration_2017["tip_movi"] == "Entrada")]
salida_migration_2017 = migration_2017[(migration_2017["tip_movi"] == "Salida")]

count_entrada = len(entrada_migration_2017.index)
count_salida = len(salida_migration_2017.index)
neto_migration_2017 = count_entrada - count_salida
print("Entrada:", count_entrada)
print("Salida:", count_salida)
print("Diferencia neta:", neto_migration_2017)
x = np.arange(2)

fig, ax = plt.subplots()
plt.bar(x, [count_entrada, count_salida])
plt.xticks(x, ('Entrada', 'Salida'))
plt.ylim(-10000, 20000)
plt.show()
```
