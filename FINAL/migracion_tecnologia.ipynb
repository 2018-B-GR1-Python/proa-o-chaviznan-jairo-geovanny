```python
def plot_pie(valores, etiquetas):
    fig1, ax1 = plt.subplots()
    ax1.pie(valores, labels=etiquetas, autopct='%1.1f%%',
            shadow=True, startangle=90)
    ax1.axis('equal')
    plt.show()
```

```python
def plot_bar(items_x, valores, etiquetas, limites, rotation="vertical"):
    fig, ax = plt.subplots()
    plt.bar(items_x, valores)
    plt.xticks(items_x, etiquetas)
    plt.ylim(*limites)
    plt.xticks(rotation=rotation)
    plt.show()
```

    * Respuesta a: ¿Como se ve la migración en Ecuador?

```python
entrada_migration_2017 = migration_2017[migration_2017["tip_movi"] == "Entrada"]
serie_migration_2017 = entrada_migration_2017['edad'].groupby(entrada_migration_2017['sex_migr']).agg('count')

df_agrupado_edad_genero = pd.DataFrame({'genero': serie_migration_2017.index, 'edad': serie_migration_2017.data})

print(df_agrupado_edad_genero)

fig1, ax1 = plt.subplots()
ax1.pie(df_agrupado_edad_genero['edad'], labels=df_agrupado_edad_genero['genero'], autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
entrada_migration_2017 = migration_2017[(migration_2017["tip_movi"] == "Entrada") & (migration_2017["cla_migr"] == "Permanentes")]
salida_migration_2017 = migration_2017[(migration_2017["tip_movi"] == "Salida") & (migration_2017["cla_migr"] == "Permanentes")]

count_entrada = len(entrada_migration_2017.index)
count_salida = len(salida_migration_2017.index)
neto_migration_2017 = count_entrada - count_salida
print("Diferencia neta:", neto_migration_2017)
x = np.arange(2)

fig, ax = plt.subplots()
plt.bar(x, [count_entrada, count_salida])
plt.xticks(x, ('Entrada', 'Salida'))
plt.ylim(-100, 200)
plt.show()
```

    * Respuesta a: ¿Como están las empresas ecuatorianas en tecnología?

```python
empresas_por_tipo_evaluadas = acti_2015["tipo"].groupby(acti_2015["tipo"]).agg('count')
print(empresas_por_tipo_evaluadas)
plot_pie(empresas_por_tipo_evaluadas.values, empresas_por_tipo_evaluadas.index)

empresas_tipo_innovacion = acti_2015["emp_inves"].groupby(acti_2015["emp_inves"]).agg('count')
plot_pie(empresas_tipo_innovacion.values, empresas_tipo_innovacion.index)

```

```python
# Gasto en investigacion y desarrollo
empresas_inversion_ID = acti_2015["ID_2014"].groupby(acti_2015["tipo"]).agg('sum')
print(empresas_inversion_ID)
plot_pie(empresas_inversion_ID.values, empresas_inversion_ID.index)
```

```python
plot_bar(np.arange(14), empresas_privadas_inversion_oaiact.values, empresas_privadas_inversion_oaiact.index, (0, 1000000000))
```

Las empresas de manufactura invierten mucho en tecnología, pero principalmente lo hacen en maquinaria. Aún asi tiene un indice de crecimiento mayor, por lo que puede ser una potencial área para trabajar.

```python
# Inversion por tipo de economia ciiu1 en empresas privadas
empresas_tipo_privadas = acti_2015[acti_2015["tipo"] == 'Privada']
empresas_tipo_privadas["crecimiento"] = pd.to_numeric(empresas_tipo_privadas["fexp"], errors='coerce')

empresas_privadas_inversion_oaiact= empresas_tipo_privadas["crecimiento"].groupby(empresas_tipo_privadas["ciiu1"]).agg('sum')
print(empresas_privadas_inversion_oaiact)
# plot_pie(empresas_inversion_oaiact.values, empresas_inversion_oaiact.index)
plot_bar(np.arange(len(empresas_privadas_inversion_oaiact.index)), empresas_privadas_inversion_oaiact.values, empresas_privadas_inversion_oaiact.index, (0, 200))
```
